Используя numpy и matplotlib, начнём с тренировки: постройте график функций $sin(x)$ и $cos(x)$ от $-\pi$ до $\pi$ из 100 точек c легендой и сеткой. Используйте для линий разные цвета, одна из них пусть будет пунктирной. Результат в этом и следующих заданиях на построение графиков должен получиться примерно как на картинке.
```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = data.drop('credit_risk_good', 1).values
y = data['credit_risk_good'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=10)
print(len(X_train), len(X_test))
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
```

```python
good_or_bad = log_reg.predict(X_test)
accuracy = round(accuracy_score(y_test, good_or_bad), 3)
precision = round(precision_score(y_test, good_or_bad), 3)
recall = round(recall_score(y_test, good_or_bad), 3)

print(f'Accuracy: {accuracy}, Precision: {precision}, Recall: {recall} of logistic regression classifier on test set')
```

```python
conf_matrix_1 = pd.DataFrame(confusion_matrix(y_test, good_or_bad), index = ['actial 0','actual 1'], columns = ['predicted 0','predicted 1'])
display(conf_matrix_1)
```

Логическая регрессия достаточно неплохо предсказвает вероятность принадлежности к группе. 
Особенно высокий recall, то есть способность классификатора находить все положительные образцы, 
но положительных оброзцов просто больше, поэтому данная метрика не особо интерересна. Также у классификатора достаточно высокая точность и accuracy.
